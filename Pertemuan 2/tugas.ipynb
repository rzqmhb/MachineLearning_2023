# Langkah 0 - Import Library

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder, StandardScaler
```

# Langkah 2 - Read Data, Lihat Informasi Data

```python
df = pd.read_csv('wbc.csv')
df.head()
```

```python
df.info()
```

# Langkah 3 - Slicing Data

```python
df.drop(df.columns[[0, 32]], axis=1, inplace=True)
df.head()
```

# Langkah 4 - Encoding Kolom Diagnosis

```python
le = LabelEncoder()
df['diagnosis'] = le.fit_transform(df['diagnosis'])
df.head()
```

```python
df.info()
```

# Langkah 5 - Standarisasi Data Numerik

```python
diagnosis = df['diagnosis']
std = StandardScaler()
df = pd.DataFrame(std.fit_transform(df), columns=df.columns)
df.drop('diagnosis', axis=1, inplace=True)
df = pd.concat([diagnosis, df], axis=1)
df.head()
```

# Langkah 6 - Stratified Split Data

```python
from sklearn.model_selection import train_test_split
df_train, df_test = train_test_split(df, test_size=0.2, random_state=0, stratify=df['diagnosis'])
print('Diagnosis : \n 1: Malignant \n 2: Benign')
print(f'Jumlah label data train:\n{df_train.diagnosis.value_counts()}')
print(f'Jumlah label data test:\n{df_test.diagnosis.value_counts()}')
```
